This is the code used to build the dataframe of all the tweets.
We used this code to build the dataframe of tweets with 100% of agreement, 75% of agreement and of m5s.
In order to build different dataframes, we need to run it each time calling different folders where are contained the respective tweets. 

```python
import pandas as pd
import os
import csv
col_names = ['_label',"tweets"]
grand_df_left = pd.DataFrame(columns = col_names)


directory2 = 'Left75'
for file in os.listdir(directory2):
    if file.endswith(".csv"):
        filedir = directory2 + "/"+file
        df = pd.read_csv(filedir)
        length,b = df.shape
        texts = ' '
        for i in range(length):
            texts += ' ' + df.iat[i,5]
        grand_df_left = grand_df_left.append({
             "_label": "label_Left",
             "tweets":  texts
              }, ignore_index=True)


grand_df_left['tweets']= grand_df_left['tweets'].replace('\n',' ', regex=True).replace('\t',' ', regex=True)
```

```python
col_names = ['_label',"tweets"]
grand_df_right = pd.DataFrame(columns = col_names)

directory2 = 'Right75'
for file in os.listdir(directory2):
    if file.endswith(".csv"):
        filedir = directory2 + "/"+file
        df = pd.read_csv(filedir)
        length,b = df.shape
        texts = ' '
        for i in range(length):
            texts += ' ' + df.iat[i,5]
        grand_df_right = grand_df_right.append({
             "_label": "label_Right",
             "tweets":  texts
              }, ignore_index=True)


grand_df_right['tweets']= grand_df_right['tweets'].replace('\n',' ', regex=True).replace('\t',' ', regex=True)
```

```python
grand_df_right.head()
```

```python
#MERGING THE 2 DATAFRAMES
col_names = ['_label',"tweets"]
base_df = pd.DataFrame(columns = col_names)

n = 1
for i in range(1100):
    base_df = base_df.append({
             "_label": "label_Right",
             "tweets":  grand_df_right.iat[i,1]
              }, ignore_index=True)
    base_df = base_df.append({
             "_label": "label_Left",
             "tweets":  grand_df_left.iat[i,1]
              }, ignore_index=True)
```
